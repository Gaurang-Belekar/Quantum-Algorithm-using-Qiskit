```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
#In this circuit, we need to find state |101> and |110> state 
#So as per the algorithm dictates, we need and winner state |w> = 1/root(2)*(|101>+|110)

oracle = QuantumCircuit(3, name ='Oracle')
oracle.cz(0, 2)
oracle.cz(1, 2)
oracle.to_gate()


oracle.draw(output = 'mpl')
```

```python
# Now Creating Diffuser
# This will be do the reflection as well as amplitude amplification
```

```python
diffuser = QuantumCircuit(3,name = 'Diffuser')
```

```python
#
diffuser.h([0,1,2])
diffuser.x([0,1,2])

# Do multi-controlled-Z gate
diffuser.h(2)
diffuser.mct([0,1],2)
diffuser.h(2)



#
diffuser.x([0,1,2])
diffuser.h([0,1,2])

diffuser.to_gate()
diffuser.draw(output = 'mpl')
```

```python
# Creating Qunatum circuit and importing our 3-qubit data, and search our oracle parameter

qc = QuantumCircuit(3,3)

#Creating data set of 3-qubit system
qc.h([0,1,2])

qc.append(oracle,[0,1,2])
qc.append(diffuser,[0,1,2])

qc.measure([0,1,2],[0,1,2])

qc.draw(output = 'mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
result = execute(qc, backend, shots = 1024).result()
counts = result.get_counts()
plot_histogram(counts)
```
