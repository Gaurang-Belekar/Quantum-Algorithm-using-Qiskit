```python
from qiskit import *
from qiskit.visualization import plot_histogram
%matplotlib inline
```

```python
#Creating a circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(2,2)

#by default in qiskit quantum registers are at |0>
# To get our wanted input we will have to apply not gate to q1

qc.x(1)

qc.barrier()
```

```python
# To look at the input
qc.draw(output = 'mpl')
```

```python
#Now we will entangle states by creating superposition of all state of q0 and making all of its superpsotion be conrolling the q1

qc.h(0)
qc.cx(0,1)

qc.barrier()

```

```python
#Storing the output in classical bits
qc.measure([0,1],[0,1])
```

qc.draw()

```python
qc.draw(output ='mpl')
```

```python
#Executing and plotting the result
backend = Aer.get_backend('qasm_simulator') 
results=execute(qc,backend,shots=1024).result()  
count = results.get_counts() 
plot_histogram(count)
```
