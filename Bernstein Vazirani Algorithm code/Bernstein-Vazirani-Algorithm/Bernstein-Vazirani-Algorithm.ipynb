```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secretNumber = '101001'
```

```python
circuit = QuantumCircuit(6+1,6) #number of bits in secret number + 1 qubit and 6 classical bit to measure
circuit.h([0,1,2,3,4,5])
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.x(6)
circuit.h(6)
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
```

```python
# Here we are receving the secret number
```

```python
# Now to create a box for guessing the secret number
#So for every 1 input, there should be CX gate for the qubit with 1 and the last qubit.
```

```python
circuit.cx(5,6) # the first bit of secret number is stored as the 6th Qubit in LIFO formart
```

```python
circuit.cx(3,6)
circuit.cx(0,6)
```

```python
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
```

```python
# Now the above oracle has guessed the secret Number
```

```python
circuit.h([0,1,2,3,4,5])
circuit.barrier()
```

```python
# Now to guess what the Quantum Computer has measured
```
